```python
'''

3.write a python function to multiply numbers in the list[77,87,86]

''
```

```python
def multiply_numbers(numbers):
    result = 1
    for num in numbers:
        result *= num
    return result

my_list =[77,87,86]
list1 = multiply_numbers(my_list)
print(list1)
```
