```python
import numpy as np
```

```python
def resiprocal(arr):
  ans = []
  for i in arr:
    ans.append(1/i)
  return np.array(ans)


resiprocal(np.array([3,4,5,6]))
```

```python
big_arr = np.random.randint(1,100,size = 1000000)
```

```python
%timeit resiprocal(big_arr)
```
