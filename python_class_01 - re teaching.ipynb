```python
#square of a number
n = input("Maximul number?") #Block 1
n = int(n) + 1 #Block1
print (n) #Block1
square = 0
for a in range(0,n): #Block1
  for b in range(a,n): #Block2
    square = a**2 + b**2 # logic under block2
  #Block 2 ends
  if square > 10: #block 3
    print (square) #logic inside block 3
```
